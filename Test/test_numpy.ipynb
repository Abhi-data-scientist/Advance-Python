```python
# 4. What is horizontal and vertical stacking explain with example
```

```python
# horizontal stacking ->  It adds 2 or more numpy array horizontally
a = np.arange(1, 10).reshape(3,3)
b = np.arange(11, 20).reshape(3,3)
np.hstack((a,b))
```

```python
# horizontal stacking ->  It adds 2 or more numpy array Vertically
a = np.arange(1, 10).reshape(3,3)
b = np.arange(11, 20).reshape(3,3)
np.vstack((a,b))
```

```python
# 5. What is view vs copy with example
```

```python
# View => in this a variable creates a copy of original array
#       -> it takes the reference of it
#       -> if i change in new array then it reflect in original array
a = np.arange(1,11)
print(a)
b = a
b[0] = 10
print(a)
```

```python
# copy => in this a variable creates a different copy of original array
#       -> it does not takes the of it
#       -> if i change in new array then it does not reflect in original array
a = np.arange(1,11)
print(a)
b = a.copy()
b[0] = 10
print(a)
print(b)
```

```python
# 7. You have to create an array in range 1 to 10 and then you have to reshape it 3*3 and
# then perform total sum, row sum ,column sum
```

```python
a = np.arange(1,10).reshape(3,3)
print(a)
print('Total sum : ', np.sum(a))
print('Row sum : ', np.sum(a, axis=1))
print('Column sum : ', np.sum(a, axis=0))
```
