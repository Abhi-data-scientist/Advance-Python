```python
# 5 How do you handle missing values using pandas
```

```python
# first we find the the null values using isnull().sum()
# second if there any null value present then
# we fill it by mean or meadian
# If there is many null values in a column then we drop that column
```

```python
df.isnull().sum() # There are 10 null values in fever column
```

```python
df['fever'].fillna(df['fever'].mean(), inplace=True)
```

```python
# 7 What is group by perform operation using group by
```

```python
# groupby =>
# In pandas groupby is a function by which we can group our data
# and then we can apply aggregation function on them
```

```python
np.round(df.groupby('city')['age'].mean(), 0)
```

```python
# 9. How to rename a column
df.rename(columns={'age':'Age'}).head(3)
```

```python
# 10 if you have two subdataframes and a column is common
# now you have toh merge these two dataframes into a singal dataframe
df1 = df.groupby('city')['age'].mean().reset_index(name='avg_age')
df2 = df.groupby('city')['age'].mean().reset_index(name='avg_fever')
pd.merge(df1, df2, on = 'city')
```

```python
# 11 You have to update the values
# from 10 to 15 index number values of all columns

# df.iloc[10:16, :] = 'Undefined'
```

```python
# 12 you have to convert categorical data into numerical data
df.head()
df['gender'] = df['gender'].map({'Male':1, 'Female':0})
df['cough'] = df['cough'].map({'Mild':0, 'Strong':1})
df['has_covid'] = df['has_covid'].map({'Yes':1, 'No':0})
df['city'] = df['city'].map({'Kolkata':0, 'Bangalore':1, 'Delhi':2, 'Mumbai':3})
```
