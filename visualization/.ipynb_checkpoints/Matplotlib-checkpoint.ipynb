```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
x = np.linspace(0,50,10)
y = x**2
```

```python
plt.plot(x, y)
plt.xlabel('x_value')
plt.ylabel('y - square of x')
plt.title('Square Graph')
plt.tight_layout()
plt.show()
```

```python
plt.subplot(2, 2,1)
plt.title('Sqare')
plt.plot(x,y)

plt.subplot(2, 2,2)
plt.title('Square root')
plt.plot(x,x**(1/2))

plt.subplot(2, 2,3)
plt.title('Sine')
plt.plot(x,np.sin(x))

plt.subplot(2, 2,4)
plt.title('Cos')
plt.plot(x,np.cos(x))

plt.tight_layout()
plt.show()
```
