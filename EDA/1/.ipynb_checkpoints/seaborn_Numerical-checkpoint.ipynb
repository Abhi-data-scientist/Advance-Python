```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("tips - tips.csv")
df.head(3)
```

```python
# Distplot / Histplot 
```

```python
sns.histplot(data=df, x='total_bill', bins=20, kde=True)
```

```python
# jointplot
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
```

```python
# pairplot 
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
# Scatter plot 
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df, hue='sex', size='size')
```

```python
# Box plot 
```

```python
sns.boxplot(x=df['total_bill'])
```
