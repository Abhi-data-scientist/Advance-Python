```python
# load the dataset and display first 10 rows
df = pd.read_csv('/content/Attrition - Attrition.csv')
df.head(10)
```

```python
# find the total number of rows & columns
df.shape
```

```python
# find list of all column name
df.columns
```

```python
#2. count unique values in department
df['Department'].value_counts()
```

```python
#3. find number of employees who left means attrition = yes.
df[df['Attrition']=='Yes'].shape[0]
```

```python
#4. filter employees whose age is greaton than 40
df[df['Age']>40]
```

```python
#5. select ony these columns age, department, monthly income.
df[['Age', 'Department', 'MonthlyIncome']]
```

```python
#6. find employees with overtime is equal to yes
df[df['OverTime']=='Yes']
```

```python
# 7 Add a new column -> income after hype -> monthly income  + (monthly income * percent sallary hike % 100)
df['Income_after_hype'] = df['MonthlyIncome'] + (df['MonthlyIncome']*df['PercentSalaryHike']/100)
df.head(3)
```

```python
#8. rename column monthly income to sallary
df.rename(columns={'MonthlyIncome' : 'Salary'})
```

```python
#9. find average monthly income per department.
df.groupby('Department')['MonthlyIncome'].mean().reset_index(name='avg_monthly_income')
```

```python
#10. find average age per job role
df.groupby('JobRole')['Age'].mean().reset_index(name='avg_age')
```

```python
#11. find minimum distance from home per department
df.groupby('Department')['DistanceFromHome'].min().reset_index(name='minimum_distance_from_home')
```

```python
#12.  find total employees per education field.
df['EducationField'].value_counts().reset_index()
```

```python
#13. find employees count grouped by marital status & attrition
df.groupby(['MaritalStatus', 'Attrition']).size().reset_index(name='emp_counts')
```
