```python
 # Unique() ---->  It will return 3 arrays 
 #    arr -> return original values of array 
 #    return_index -> True ==> return index value of unique numbers
 #    return_counts -> True ==> It will return total frequency of each unique value 
```

```python
a = np.array([1,2,3,2,2,2,3,3,2,4,5])
a
```

```python
np.unique(a, return_index=True, return_counts=True)
```

```python
# hstack() and vstack() 

# hstack() -> it will add 2 or more arrays in horizontally
# hstack() -> it will add 2 or more arrays in vertically
```

```python
a = np.arange(1,5).reshape(2,2)
b = np.arange(5,9).reshape(2,2)
c = np.arange(9,13).reshape(2,2)
```

```python
np.hstack((a,b,c))
```

```python
np.vstack((a,b,c))
```
