```python
import numpy as np 
```

```python
a = [1,2,3]
a*3
```

```python
a = np.array([1,2,3])
a*3
```

```python
a = [1,2,3]
b = np.array(a) 
b
```

```python
print(type(a)) 
print(type(b))
```

```python
a = [[1,2,3],[4,5,6],[7,8,9]]
b = np.array(a) 
b
```

```python
print('Total Shape : ',b.shape)      ## shape -> row,column
print('Total Elements : ',b.size)    ## size -> row*column 
print('Total Dimensions : ',b.ndim)  ## total dimension = 2
```

### Pre-Built functions of Numpy

```python
a = np.zeros(3)
a
```

```python
a = np.zeros((3,4))  ## np.zeros(shape)
a
```

```python
a = np.ones(4) 
a
```

```python
b = np.ones((2,3))
b
```

```python
a = np.ones((5,6))
print(a)
print('Shape : ', a.shape)
print('dimension : ', a.ndim) 
print('Total elements : ', a.size) 
```
