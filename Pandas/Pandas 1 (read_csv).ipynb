```python
# Pandas => 
    # -> Pandas is an open source library which is used for handle data manipulation.
```

```python
import pandas as pd 
```

```python
# Data Structure of Pandas ::: 
    # 1. Series --> It stores only values not column name and it is one-dimensional 
    # 2. DataFrame --> It store values with column name, and it is multi-dimensional 
```

```python
a = pd.Series([1,2,3,4,5,6])
a
```

```python
type(a)
```

```python
# DataFrame 
```

```python
a = {
    'Emp_id' : [1,2,3,4,5],
    'Name' : ['Abhi', 'Akanksha', 'Vikash', 'Divyansh', 'Rishi'],
    'Department' : ['IT', 'Sales', 'It', 'Sales', 'IT'],  
    'Salary' : [20000, 25000, 40000, 30000, 35000]
}
```

```python
type(a)
```

```python
df = pd.DataFrame(a)
df
```

```python
# Export this data into csv 
```

```python
df.to_csv('Emp_data.csv')
```

```python
# Export this data into Excel 
```

```python
df.to_excel('New_Emp_data.xlsx')
```
