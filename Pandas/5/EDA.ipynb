## Parts of EDA 
1. Univariate Analysis -> Analysis on a singal independent column
2. Bivariate Analysis -> Analysis on Two columns
3. Multivairate Analysis -> Analysis on more than 2 columns

```python
# Data Types 
# 1. Numerical data -> continous data -> Age, year, month, days, height, weight 
# 2. Categorical data -> descrete data -> total number of employees, marital status, student result
```

```python
import numpy as np 
import pandas as pd  
```

```python
import matplotlib.pyplot as plt  # visualization Library
import seaborn as sns  # matplotlib's updated version is seaborn
```

```python
df = pd.read_csv('titanic - titanic.csv')
```

```python
df.head()
```

### Univariate Analysis

```python
df.columns
```

```python
df['Survived'].value_counts()
```

```python
sns.countplot(x=df['Survived'])
```

```python
# If we want to see actual values then we use value_counts(). 

df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
# If we want to find out percentage then use piechart.abs 
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
# If we have numerical data then we use histogram 
# bcz it find the distribution
```

```python
plt.hist(x=df['Age'])
plt.show()
```

```python
# Boxplot -> use for find outliers

# 1. Lower fence 
# 2. 25% data 
# 3. IQR (Inter quartile range) (75%-25%)
# 4. 75% data 
# 5. upper fence
```

```python
# x = 1,2,3,4,5 
# mean = (1+2+3+4+5)/5 = 15/5 = 3 

# x = 1,2,3,4,5,100   #here 1000
# mean = 115/6=19.6
```

```python
sns.boxplot(x=df['Age'])
```
