```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Churn_Modelling - Churn_Modelling.csv')
```

```python
df.head(5)
```

```python
# 1. Find the total number of customers in each country.
```

```python
df.groupby(['Geography']).size().reset_index(name='counts')
```

```python
# 2. Find the average credit score of customers based on gender.
```

```python
df.groupby('Gender')['CreditScore'].mean().reset_index()
```

```python
# 3. Find the total number of customers who exited and customers who stayed.
```

```python
df.groupby('Exited').size().reset_index(name='count')
```

```python
# 4. Find the average account balance in each country.
```

```python
df.groupby('Geography')['Balance'].mean().reset_index(name='Avg_Balance')
```

```python
# 5. Find which age has the highest number of customers who exited the bank.
```

```python
df[df['Exited']==1].groupby('Age').size().reset_index(name='counts').sort_values(by='counts', ascending=False).iloc[:1, :]
```

```python
# 6. Find the average estimated salary of active and inactive members.
```

```python
df.groupby('IsActiveMember')['EstimatedSalary'].mean().reset_index(name='Avg_EstimatedSalary')
```

```python
# 7. Find the churn count based on the number of products customers are using.
```

```python
df[df['Exited']==1].groupby('NumOfProducts').size()
```

```python
# 8. Create a separate DataFrame containing country and region information, merge it with the
# main dataset using pd.merge(), and then find the average balance and average estimated
# salary by region.
```

```python
region_df = pd.DataFrame({
    'Geography': ['France', 'Germany', 'Spain'],
    'Region': ['Western Europe', 'Central Europe', 'Southern Europe']
})
region_df
```

```python
df_merge = pd.merge(region_df, df, on='Geography')
```

```python
df_merge.groupby('Region')['Balance'].mean('Balance').reset_index(name='Avg_balance')
```

```python
df_merge.groupby('Region')['EstimatedSalary'].mean('EstimatedSalary').reset_index(name='Avg_EstimatedSalary')
```
